```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
```

```python
df = pd.read_csv('sonar data.csv')
df.rename(columns={'R':'Label'}, inplace=True)
df.head()
```

```python
df.describe()
```

```python
df.groupby('Label').mean()
```

```python
X = df.drop(columns='Label', axis=1)
y = df['Label']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42, stratify=y)
```

```python
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
sm = SMOTE(random_state=42)
X_train_bal, y_train_bal = sm.fit_resample(X_train, y_train)
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train_bal, y_train_bal)
```

```python
y_pred = model.predict(X_test)
```

```python
class_rep = classification_report(y_test, y_pred)
acc_scr = accuracy_score(y_test, y_pred)

print(f'Accuracy Score: {acc_scr}')
print(f'Classification Report: {class_rep}')
```

```python
import numpy as np

input_data = list(map(int,input().split(',')))

input_data = np.asarray(input_data)

input_shaped = input_data.reshape(1,-1)

predicted = model.predict(input_shaped)

if predicted[0] == 'R':
    print('The object is a rock.')
else:
    print('The object is a mine.')
```
