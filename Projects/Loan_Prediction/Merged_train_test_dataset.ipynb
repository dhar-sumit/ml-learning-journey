```python
import pandas as pd
```

```python
train_csv = pd.read_csv('train.csv')
test_csv = pd.read_csv('test.csv')
```

```python
combined_csv = pd.concat([train_csv, test_csv], ignore_index=True)
combined_csv
```

```python
shuffled_csv = combined_csv.sample(frac=1, random_state=42).reset_index(drop=True)
shuffled_csv
```

```python
shuffled_csv.to_csv('loan.csv', index=False)
```
