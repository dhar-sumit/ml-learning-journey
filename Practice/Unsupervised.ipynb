```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cols = ['area', 'perimeter', 'compactness', 'length', 'width', 'asymmetry', 'groove', 'class']
df = pd.read_csv('seeds_dataset.txt', sep='\s+', names=cols)
df.head()
```

```python
for i in range(len(cols)-1):
    for j in range(i+1, len(cols)-1):
        x_label = cols[i]
        y_label = cols[j]
        sns.scatterplot(x=x_label, y=y_label, data=df, hue='class')
        plt.show()
```

# CLustering

```python
from sklearn.cluster import KMeans
```

```python
x = 'compactness'
y = 'asymmetry'
X = df[[x, y]].values
```

```python
k_means = KMeans(n_clusters=3)
k_means.fit(X)
```

```python
clusters = k_means.labels_
clusters
```

```python
cluster_df = pd.DataFrame(np.hstack((X, clusters.reshape(-1, 1))), columns=[x, y, 'class'])
cluster_df.head()
```

```python
# K Means classes
sns.scatterplot(data=cluster_df, x=x, y=y, hue='class')
plt.plot()
```

```python
# Original classes
sns.scatterplot(data=df, x=x, y=y, hue='class')
plt.plot()
```

# Higher Dimensions

```python
X = df[cols[:-1]].values
```

```python
k_means = KMeans(n_clusters=3).fit(X)
clusters = k_means.labels_
cluster_df = pd.DataFrame(np.hstack((X, clusters.reshape(-1, 1))), columns=df.columns)
```

```python
# K Means classes
sns.scatterplot(data=cluster_df, x=x, y=y, hue='class')
plt.plot()
```

```python
# Original classes
sns.scatterplot(data=df, x=x, y=y, hue='class')
plt.plot()
```
